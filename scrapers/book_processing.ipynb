```python
import pandas as pd
from pathlib import Path
```

```python
# Get current files in "out" directory
files = Path("./out").glob("*")

# Read in csv files
df = pd.concat(map(pd.read_csv, files), ignore_index=True)

print(f"Size of df: {len(df)}")
df.head()
```

```python
# Remove duplicates from the dataset
df_clean = df.drop_duplicates(subset=["isbn"])

print(f"Size of df: {len(df_clean)}")
df_clean.head()
```
